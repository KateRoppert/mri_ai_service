# 1. Split train, test subsets from BraTS-MET Training dataset

```python
import os, shutil
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
DST_DIR = "/media/storage/luu/nnUNet_raw/Dataset555_BraTSMET"
SRC_DIR = "/media/storage/luu/nnUNet_raw/Dataset111_Meta"

dst_imagesTr = os.path.join(DST_DIR, "imagesTr")
dst_labelsTr = os.path.join(DST_DIR, "labelsTr")
dst_imagesTs = os.path.join(DST_DIR, "imagesTs")
dst_labelsTs = os.path.join(DST_DIR, "labelsTs")

os.makedirs(dst_imagesTr, exist_ok=True)
os.makedirs(dst_labelsTr, exist_ok=True)
os.makedirs(dst_imagesTs, exist_ok=True)
os.makedirs(dst_labelsTs, exist_ok=True)

src_imagesTr = os.path.join(SRC_DIR, "imagesTr")
src_labelsTr = os.path.join(SRC_DIR, "labelsTr")
```

```python
src_ids = os.listdir(src_labelsTr)
src_ids = [id_.replace(".nii.gz", "") for id_ in src_ids]
len(src_ids)
```

```python
patient_ids = np.unique([src_id.split("-")[2] for src_id in src_ids])
len(src_ids), len(patient_ids)
```

```python
# Split by patient level
train_patients, test_patients = train_test_split(patient_ids, test_size=0.1, random_state=42)
assert len(train_patients) + len(test_patients) == len(patient_ids)
```

```python
# Collect train and test sets of ids by session level
train_ids = [
    src_id
    for src_id in src_ids
    if any(patient in src_id for patient in train_patients)
]

test_ids = [
    src_id for src_id in src_ids if src_id not in train_ids
]

assert len(train_ids) == len(set(train_ids))
assert len(test_ids) == len(set(test_ids))
assert len(train_ids) + len(test_ids) == len(src_ids)
```

```python
train_ids[:2], test_ids[:2]
```

```python
# Copy to nnUNet_raw

# Copy training images and labels
train_image_count = 0
train_label_count = 0
for id_ in train_ids:
    # Copy images
    fns = [f"{id_}_000{i}.nii.gz" for i in range(4)]

    src_paths = [os.path.join(src_imagesTr, fn) for fn in fns]
    for path in src_paths:
        assert os.path.isfile(path)
    
    dst_paths = [os.path.join(dst_imagesTr, fn) for fn in fns]

    for src_path, dst_path in zip(src_paths, dst_paths):
        try:
            print(f"Copy from {src_path} to {dst_path}")
            shutil.copy2(src_path, dst_path)
        except Exception as err:
            print(f"Error copying {src_path}, err: {err}")
    train_image_count +=1

    # Copy labels
    fn = f"{id_}.nii.gz"
    src_path = os.path.join(src_labelsTr, fn)
    dst_path = os.path.join(dst_labelsTr, fn)
    try:
        print(f"Copy from {src_path} to {dst_path}")
        shutil.copy2(src_path, dst_path)
    except Exception as err:
        print(f"Error copying {src_path}, err: {err}")
    train_label_count += 1


# Copy testing images and labels
test_image_count = 0
test_label_count = 0
for id_ in test_ids:
    # Copy images
    fns = [f"{id_}_000{i}.nii.gz" for i in range(4)]

    src_paths = [os.path.join(src_imagesTr, fn) for fn in fns]
    for path in src_paths:
        assert os.path.isfile(path)
    
    dst_paths = [os.path.join(dst_imagesTs, fn) for fn in fns]

    for src_path, dst_path in zip(src_paths, dst_paths):
        try:
            print(f"Copy from {src_path} to {dst_path}")
            shutil.copy2(src_path, dst_path)
        except Exception as err:
            print(f"Error copying {src_path}, err: {err}")
    test_image_count +=1

    # Copy labels
    fn = f"{id_}.nii.gz"
    src_path = os.path.join(src_labelsTr, fn)
    dst_path = os.path.join(dst_labelsTs, fn)
    try:
        print(f"Copy from {src_path} to {dst_path}")
        shutil.copy2(src_path, dst_path)
    except Exception as err:
        print(f"Error copying {src_path}, err: {err}")
    test_label_count += 1

train_image_count, train_label_count, test_image_count, test_label_count
```
